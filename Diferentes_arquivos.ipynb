# Desafio csv

_> objetivo: ler o arquivo csv a partir da 4 linha, já que é a que contém o cabeçalho, excluir as últimas 9 linhas.

```python
import pandas as pd
dadosimportados= 'dados_sus.csv'
dadoslidos= pd.read_csv(dadosimportados, encoding= 'ISO-8859-1', skiprows= 3, sep= ';')
dadoslidos
```

```python
dadoslidos.fillna(0, inplace= False)
```

```python
descobrindo_indices_para_remover= dadoslidos.tail(9).index
descobrindo_indices_para_remover
```

```python
dadoslidos.drop(descobrindo_indices_para_remover, inplace= False)
```
