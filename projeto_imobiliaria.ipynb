# Tratando e filtrando os dados

## Lidando com dados nulos

```python
df.info()
```

```python
nan= df.isna()
nan.sum()
```

```python
linhascomnulos= df[nan.any(axis=1)]
linhascomnulos
```

```python
df= df.loc[~df.index.isin(linhascomnulos.index)]
df
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



```python
filtro1= df.query('Tipo=="Apartamento" & Quartos== 1 & Valor < 1200' )
filtro1
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
filtro2= df.query('Tipo == "Apartamento" & Quartos >= 2 & Valor < 3000 & Area > 70')
filtro2
```

```python
filtro2['Area'].min()
```

## Salvando os dados

```python
filtro1.to_csv('aptos_1quarto.csv')
```
