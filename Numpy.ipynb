# Laranja e toranjas

OUTRO DESAFIO


```python
import numpy as np
```

```python
url= 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
```

```python
dados= np.loadtxt(url, delimiter = ',', usecols= np.arange(1, 6, 1),  skiprows= 1 )
print (dados)
```

```python
np.arange(1, 6, 1)
```

```python
dados
```

```python
diam_laranja= dados[:5000, 0]
diam_laranja
```

```python
peso_la=dados[:5000, 1]
peso_la
```

```python
diam_toranja= dados[5000:, 0]
diam_toranja
```

```python
peso_toranja= dados[5000:, 1]
peso_toranja
```

```python
import matplotlib.pyplot as plt
```

```python
x= diam_laranja
y= 6*x+120
```

```python
plt.plot(diam_laranja,peso_la)
plt.plot(x, y)
plt.title('diametro x peso laranjas')
plt.xlabel('diametro')
plt.ylabel('peso')
plt.show()
```

```python
np.linalg.norm(peso_la - y)
```

```python
Y= peso_la
X= diam_laranja
n= np.size(peso_la)
```

```python
a= (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - (np.sum(X)**2))
```

```python
b= np.mean(Y) - a * np.mean(X)
```

```python
y= a*X+b
```

```python
np.linalg.norm(peso_la - y)
```

```python
plt.plot(diam_laranja,peso_la)
plt.plot(x, y)
plt.plot(7.11, 7.11*a+b,'*b')
plt.title('diametro x peso laranjas')
plt.xlabel('diametro')
plt.ylabel('peso')
plt.show()
```

```python
# Calcular os coeficientes da regressão linear
a, b = np.polyfit(x, y, 1)

# Calcular o R²
y_mean = np.mean(y)
SST = np.sum((y - y_mean)**2)
y_pred = a * x + b
SSR = np.sum((y_pred - y_mean)**2)
SSE = np.sum((y - y_pred)**2)
R2 = SSR / SST

print(f"Coeficiente de determinação (R²): {R2:.2f}")
```

```python
# Calcular os coeficientes da regressão linear
a, b = np.polyfit(x, y, 1)

# Calcular o R²
y_mean = np.mean(y)
SST = np.sum((y - y_mean)**2)
y_pred = a * x + b
SSR = np.sum((y_pred - y_mean)**2)
SSE = np.sum((y - y_pred)**2)
R2 = SSR / SST

print(f"Coeficiente de determinação (R²): {R2:.2f}")
```
